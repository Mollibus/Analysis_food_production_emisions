```python
df = pd.read_csv("Food_Production.csv")
df1 = pd.read_csv("ghg-kcal-poore.csv")
df2 = pd.read_csv("land-use-kcal-poore.csv")
df3 = pd.read_csv("land-use-per-kg-poore.csv")
FAO = pd.read_csv("FAO.csv",encoding = "ISO-8859-1")
```

We can already see that in the main dataset there are many missing value, however the columns related to the food production suplly chain appear to be clean, therefore I'm going to create a subset in order to separate them from the uncleaned ones.



```python
food_production = df[['Food product', 'Land use change', 'Animal Feed', 'Farm', 'Processing',
       'Transport', 'Packging', 'Retail', 'Total_emissions']]
```

```python
food_production.head()
```

Considering that I want to merge the additional dataset to overrite missing values I need to check wether the number of rows of the datasets it's the same

```python
print("Datasets shape","\n","df:",df.shape,"\n","df1:",df1.shape,"\n","df2:",df2.shape,"\n","df3:",df3.shape)
```

main dataset has 43 rows and others have 38 rows only, therefore I'll use a for loop to see what kind of food product is missing in other datasets

```python
for food in df["Food product"]: 
    if not food in list(df1["Entity"]):
        print(food)
    elif not food in list(df2["Entity"]): 
        print(food)
    elif not food in list(df3["Entity"]): 
        print(food)
```

The food products reported above are not going to be considered in the analysis of the dataset that I'm going to create below

```python
food_data = df[['Food product','Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)',
       'Eutrophying emissions per kilogram (gPO₄eq per kilogram)', 'Scarcity-weighted water use per kilogram (liters per kilogram)',
                "Scarcity-weighted water use per 1000kcal (liters per 1000 kilocalories)"]]

food_data = food_data.merge(df1, left_on='Food product', right_on='Entity')
food_data = food_data.merge(df2, left_on='Food product', right_on='Entity')
food_data = food_data.merge(df3, left_on='Food product', right_on='Entity')

food_data.drop(["Entity_x","Code_x","Year_x","Entity_y","Code_y","Year_y","Entity","Code","Year"], axis=1, inplace=True)

food_data.head()
```

```python
food_data.isnull().sum()
```

As there are still missing values I'm going to drop the rows with NaN values during the corresponding variable-related analysis

# Data Analysis

```python
tot_emission = food_production.sort_values("Total_emissions", ascending= False)
emission_mean = tot_emission["Total_emissions"].mean()
quantile_25 = tot_emission["Total_emissions"].quantile(q=0.25)
quantile_75 = tot_emission["Total_emissions"].quantile(q=0.75)

plt.figure(figsize=(10,10))
plt.title("Greenhouse emmisions (Kg CO2 - equivalents per kg product)")
sns.barplot(x ="Total_emissions", y= "Food product", data = tot_emission)
plt.axvline(x=quantile_25,ymin = 0, ymax=1,color=("red"),linestyle="--" )
plt.axvline(x=quantile_75,ymin = 0, ymax=1,color=("black"),linestyle=":" )
plt.show()
```

```python
quartile_75 = tot_emission["Total_emissions"] > quantile_75
quartile_25 = tot_emission["Total_emissions"] < quantile_25
medium = (tot_emission["Total_emissions"] <= quantile_75 ) & (tot_emission["Total_emissions"] >= quantile_25)
```

```python
high_emissions = tot_emission["Food product"][quartile_75]
average_emissions = tot_emission["Food product"][medium]
low_emissions = tot_emission["Food product"][quartile_25]
```

```python
average_emissions
```

```python
# we can have a better understanding of food impact on different stage of live cycle in food production

fig = pd.DataFrame(tot_emission.sort_values("Total_emissions").iloc[:,:-1])
fig.set_index('Food product').plot(kind='barh', stacked=True, figsize=(15,10))
plt.title("Emisions of Co2 during food production life cycle")
plt.legend(loc='lower center', ncol=7,fontsize=11)
plt.show()
```

```python
percentage = pd.DataFrame(tot_emission.iloc[:,1:-1].sum())
percentage.apply(lambda x:x/x.sum()*100).round(2)
```

```python

plt.figure(figsize=(12,12))
eutrophying_kg = food_data.dropna().sort_values('Eutrophying emissions per kilogram (gPO₄eq per kilogram)',ascending = False)
plt.subplot(221)
sns.barplot(x='Eutrophying emissions per kilogram (gPO₄eq per kilogram)', y= "Food product", data= eutrophying_kg)
plt.title("Eutrophying emissions per kilogram")
plt.xlabel("gPO₄eq per kilogram") 


scarcity_kg = food_data.dropna().sort_values("Scarcity-weighted water use per kilogram (liters per kilogram)",ascending = False)
plt.subplot(223)
sns.barplot(x="Scarcity-weighted water use per kilogram (liters per kilogram)", y= "Food product", data= scarcity_kg)
plt.title("Scarcity-weighted water use per kilogram")
plt.xlabel("liters per kilogram") 


landuse_kg = food_data.sort_values('Land use per kilogram (Poore & Nemecek, 2018)',ascending = False)
plt.subplot(222)
sns.barplot(x='Land use per kilogram (Poore & Nemecek, 2018)', y= "Food product", data= landuse_kg)
plt.title("Land use per kilogram ")
plt.xlabel("m² per kilogram") 

plt.tight_layout(pad=2.0)
plt.show()
```

```python
plt.figure(figsize=(12,12))



eutrophying_kcal = food_data.dropna().sort_values('Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)',ascending = False)
plt.subplot(221)
sns.barplot(x='Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)', y= "Food product", data= eutrophying_kcal)
plt.title("Eutrophying emissions per 1000kcal")
plt.xlabel("(gPO₄eq per 1000kcal") 

emissions_kcal = food_data.sort_values("GHG emissions per 1000kcal (Poore & Nemecek, 2018)",ascending = False)
plt.subplot(222)
sns.barplot(x="GHG emissions per 1000kcal (Poore & Nemecek, 2018)", y= "Food product", data= emissions_kcal)
plt.title("Greenhouse gas emissions per 1000kcal")
plt.xlabel("kgCO₂eq per 1000kcal") 

scarcity_kcal = food_data.dropna().sort_values('Scarcity-weighted water use per 1000kcal (liters per 1000 kilocalories)',ascending = False)
plt.subplot(223)
sns.barplot(x='Scarcity-weighted water use per 1000kcal (liters per 1000 kilocalories)', y= "Food product", data= scarcity_kcal)
plt.title("Scarcity-weighted water use per 1000kcal")
plt.xlabel("liters per 1000 kilocalories") 

landuse_kcal = food_data.sort_values('Land use per 1000kcal (Poore & Nemecek, 2018)',ascending = False)
plt.subplot(224)
sns.barplot(x='Land use per 1000kcal (Poore & Nemecek, 2018)', y= "Food product", data= landuse_kcal)
plt.title("Land use per 1000kcal")
plt.xlabel("m² per 1000kcal") 

plt.tight_layout(pad=2.0)
plt.show()
```

```python
FAO_2013 = pd.DataFrame(FAO[['Area Abbreviation', 'Area Code', 'Area', 'Item Code', 'Item',
       'Element Code', 'Element', 'Unit', 'latitude', 'longitude', 'Y2013']])
```

```python
FAO_2013.head()
```

```python
FAO_2013["Item"].value_counts()
```

```python
high = ['Bovine Meat', 'Poultry Meat','Coffee and products',
        'Pigmeat','Palm Oil','Mutton & Goat Meat']

average = ['Eggs','Rice (Milled Equivalent)','Milk - Excluding Butter','Tomatoes and products','Sunflowerseed Oil','Fish, Seafood',
         'Olive Oil','Soyabean Oil','Pulses, Other and products','Maize and products', 'Barley and products','Wheat and products',]

low = ['Potatoes and products','Nuts and products','Vegetables, Other','Bananas','Apples and products','Fruits, Other','Vegetables','Roots, Other',
                'Maize and products','Cassava and products','Sugar cane','Citrus, Other',]
```

```python
subset = pd.DataFrame(FAO_2013[FAO_2013["Item"].isin(high) | FAO_2013["Item"].isin(low) | FAO_2013["Item"].isin(average)])
```

```python
subset.head()
```

```python
subset["Emissions"] = subset["Item"]
```

```python

for n in subset["Item"]: 
    if n in high: 
        subset["Emissions"].replace({n:"high"},inplace=True)
    elif n in low: 
        subset["Emissions"].replace({n:"low"},inplace=True)
    elif n in average:
        subset["Emissions"].replace({n:"avg"},inplace=True)
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(x="Y2013",hue="Emissions",data=subset,shade="fill")
plt.xlim(right=10000)
plt.show()
```
